## DecisionTree-regresor

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
df=fetch_california_housing()
```

```python
df.keys()
```

```python
df_data=pd.DataFrame(data=df.data,columns=df.feature_names)
```

```python
df_data['price']=df.target
```

```python
df_data.head()
```

```python
sns.set(rc={'figure.figsize':(7,5)})
sns.heatmap(df_data.corr(),annot=True,vmin=-1,vmax=1,center=0)
```

```python
x=df_data.loc[:,['MedInc','Latitude','HouseAge']]
y=df_data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=35,random_state=1)

from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()

dt.fit(X_train,y_train)
y_pred= dt.predict(X_test)

from sklearn.metrics import r2_score
print("R2:",r2_score(y_test,y_pred))
```
